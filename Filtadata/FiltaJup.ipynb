```python
active_variable = filta.loc[filta["Status"] == "Active", :]
ActiveSum = active_variable["Status"].count()
ActiveSum
```

```python
active_variable = filta.loc[filta["Status"] == "Active", :]
ActiveSum = active_variable["Status"].count()

inactive_variable = filta.loc[filta["Status"] == "Inactive", :]
InactiveSum = inactive_variable["Status"].count()



Totalsum = ActiveSum + InactiveSum
#Totalsum

AP = ActiveSum / Totalsum
IAP = InactiveSum / Totalsum


types = ["Active", "Inactive"]
percents = [AP, IAP]
colors = ["lightcoral", "Gold"]
explode = (0.1, 0)

plt.title("% of Status")
plt.pie(percents, explode=explode, labels=types, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=280)
#plt.axis("equal")
plt.savefig("Images/Statuspie.png")
plt.show()
```

```python
x_axis = np.arange(len(statemeans))
tick_locations = [value for value in x_axis]
```

```python
plt.figure(figsize=(20,10))
plt.bar(x_axis, statemeans["Territories"], color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, statearray, rotation="vertical")

plt.title("Average by State")
plt.xlabel("State")
plt.ylabel("Average Franchise In State")
#plt.tight_layout()
plt.savefig("Images/avgfiltafranchise.png")
plt.show()
```

```python
Jerseyfranchise = filta.loc[filta["State"] == "New Jersey", :]
Jerseycount = Jerseyfranchise["State"].count()
NYfranchise = filta.loc[filta["State"] == "New York", :]
NYcount = NYfranchise["State"].count()
PAfranchise = filta.loc[filta["State"] == "Pennsylvania",:]
PAcount = PAfranchise["State"].count()
TotalCount = Jerseycount + NYcount + PAcount
Totalfranchise = filta["State"].nunique()

Tripercent = TotalCount / Totalfranchise
Otherpercent = (1-Tripercent)
```

```python
Active = filta.loc[filta["Status"] == "Active", :]
Total = Active["Territories"].sum()
```

```python
PAFcount = PAfranchise["Territories"].sum()
NJFcount = Jerseyfranchise["Territories"].sum()
NYFcount = NYfranchise["Territories"].sum()
Totaltricount = PAFcount + NJFcount + NYFcount

Totaltripercent = Totaltricount/Total
Otherp = 1 - Totaltripercent
```

```python
types = ["Tristate", "Other"]
percents = [Totaltripercent, Otherp]
colors = ["Blue", "Green"]
explode = (0.1, 0)

plt.title("% of Tri-State Territories")
plt.pie(percents, explode=explode, labels=types, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=280)
#plt.axis("equal")
plt.savefig("Images/Triterritorypie.png")
plt.show()
```

```python

types = ["Tristate", "Other"]
percents = [Tripercent, Otherpercent]
colors = ["Blue", "Gold"]
explode = (0.1, 0)

plt.title("% of Tri-State Franchises")
plt.pie(percents, explode=explode, labels=types, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=280)
#plt.axis("equal")
plt.savefig("Images/Tripie.png")
plt.show()
```
