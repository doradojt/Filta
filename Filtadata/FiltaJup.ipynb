```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from stats import mean, median, mode, multi_mode
```

```python
filta = pd.read_csv("filtadata.csv")
```

```python
filta.head()
```

```python
#average Territory is 1.853591  need to define how Filta gives out territories since not all are equal
avgfranchise = filta["Territories"].mean()
avgfranchise
```

```python
maxfranchise = filta["Territories"].max()
minfranchise = filta["Territories"].min()
```

```python
unique_filta_data = filta.nunique()
unique_filta_data
```

```python
#total franchise count is 181
franchisecount = filta["Territories"].count()
```

```python
#count of unique states that have atleast 1 filta franchise
statecount = filta["State"].nunique()
statecount
```

```python
#average franchise per state count
avgperstate = franchisecount/statecount
avgperstate
```

```python
status_count = filta["Status"].count()
status_count
```

```python
gbstate = filta.groupby(["State"])
statemeans = gbstate.mean()
#statemeans["Territories"]
```

```python
statearray = filta["State"].unique()
```

```python
groupedbystatus = filta.groupby(["Status"])
#groupedbystatus.count()
```

```python
x_axis = np.arange(len(statemeans))
tick_locations = [value for value in x_axis]
```

```python
plt.figure(figsize=(20,3))
plt.bar(x_axis, statemeans["Territories"], color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, statearray, rotation="vertical")

plt.title("Average by State")
plt.xlabel("State")
plt.ylabel("Average Franchise In State")
#plt.tight_layout()
plt.savefig("Images/avgfiltafranchise.png")
plt.show()
```
